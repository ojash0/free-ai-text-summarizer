```python
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization")
```

```python
text = """
Artificial Intelligence (AI) is rapidly transforming how we live and work. It powers everything from self-driving cars
to virtual assistants and is now being used in medicine, education, finance, and many other industries.
While it brings many benefits, such as improved efficiency and productivity, it also raises concerns
about job loss, privacy, and ethical use.
"""
```

```python
# Summarize the text
summary = summarizer(text, max_length=60, min_length=30, do_sample=False)

# Print the summary
print("📝 Summary:")
print(summary[0]['summary_text'])
```
